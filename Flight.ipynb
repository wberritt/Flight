# Data Cleaning/Prep

```python
df1.value_counts("PREVIOUS_AIRPORT")
```

```python
df1 = df1[df1.PREVIOUS_AIRPORT != "NONE"]
df1.shape
```

```python
series = pd.value_counts(df1.PREVIOUS_AIRPORT)
mask = (series/series.sum() * 100).lt(1)

df1['PREVIOUS_AIRPORT'] = np.where(df1['PREVIOUS_AIRPORT'].isin(series[mask].index),'Other',df1['PREVIOUS_AIRPORT'])
```

```python
df1.value_counts("PREVIOUS_AIRPORT")
```

```python
import numpy as np
conditions = [
    (df1['PRCP'] == 0),
    (df1['PRCP'] !=0)]
values = ['No Precip', 'Yes Rain']
df1['Rain'] = np.select(conditions, values)

import numpy as np
conditions = [
    (df1['SNWD'] == 0),
    (df1['SNWD'] !=0)]
values = ['No Snow Accum', 'Snow Accum']
df1['Ground Snow'] = np.select(conditions, values)

df1.head()
```

```python
df1.drop('LATITUDE', inplace=True, axis=1)
df1.drop('AIRPORT_FLIGHTS_MONTH', inplace=True, axis=1)
df1.drop('AVG_MONTHLY_PASS_AIRPORT', inplace=True, axis=1)
df1.drop('AVG_MONTHLY_PASS_AIRLINE', inplace=True, axis=1)
df1.drop('SNOW', inplace=True, axis=1)
df1.drop('PRCP', inplace=True, axis=1)
df1.drop('SNWD', inplace=True, axis=1)
```

```python
df1.drop('LONGITUDE', inplace=True, axis=1)
```

```python
DDB = pd.get_dummies(df1['DEP_BLOCK'])
DDB.head()

DCN = pd.get_dummies(df1['CARRIER_NAME'])
DCN.head()

DDA = pd.get_dummies(df1['DEPARTING_AIRPORT'])
DDA.head()

DPA = pd.get_dummies(df1['PREVIOUS_AIRPORT'])
DPA.head()

DRain = pd.get_dummies(df1['Rain'])
DSnow = pd.get_dummies(df1['Ground Snow'])

dfmodel = pd.concat([df1, DPA, DDA, DCN, DDB, DRain, DSnow], axis=1)
dfmodel.drop('DEP_BLOCK', inplace=True, axis=1)
dfmodel.drop('CARRIER_NAME', inplace=True, axis=1)
dfmodel.drop('DEPARTING_AIRPORT', inplace=True, axis=1)
dfmodel.drop('PREVIOUS_AIRPORT', inplace=True, axis=1)
dfmodel.drop('Rain', inplace=True, axis=1)
dfmodel.drop('Ground Snow', inplace=True, axis=1)

dfmodel.head()
```

# Modeling

```python
from sklearn.model_selection import train_test_split
x=dfmodel.loc[:, dfmodel.columns != 'DEP_DEL15']
y=dfmodel[['DEP_DEL15']]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
xtrain.columns
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn import decomposition, datasets
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
randf=rfc().fit(xtrain, ytrain)

randf_results = randf.score(xtest, ytest)
print('RandomF Score:')
print(randf_results)
ypredrfc = randf.predict(xtest)
rfcmatrix = confusion_matrix(ytest, ypredrfc)
print('RFCMatrix Results:')
print(rfcmatrix)
rfcreport = classification_report(ytest, ypredrfc)
print('RFC Report:')
print(rfcreport)
```

```python
dectree = dtc().fit(xtrain, ytrain)

print('Dectree Score:', dectree.score(xtest, ytest)) 
ypreddtc = dectree.predict(xtest) 
dtcmatrix = confusion_matrix(ytest, ypreddtc)
print('DTCmatrix:') 
print(confusion_matrix(ytest, ypreddtc))
print('dtcreport:') 
print(classification_report(ytest, ypreddtc))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
params = {'max_leaf_nodes': [25, 50, 75, 100], 'min_samples_split': [1000, 10000, 20000]}
grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42), params, verbose=1, cv=3)
grid_search_cv.fit(xtrain, ytrain)
```
